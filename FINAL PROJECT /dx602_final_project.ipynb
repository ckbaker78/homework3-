# 2.) Loading and Inspecting the Data

```python
# Imports 

import pandas as pd 
import matplotlib.pyplot as plt 
import sklearn 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error


game_data = pd.read_csv('game_stats_v2.csv') # loading the data 
game_data.sample(5) #sample rows 

game_data.columns # column names 
game_data.dtypes # data types 
```

# 3.) Data Cleaning

```python
game_data.isna().sum() # the number of NA values per column

game_data[game_data['home_team'].isna()] 
game_data = game_data.dropna(subset=['home_team', 'away_team']).copy()
print('sample rows:',game_data.sample(5),'\n') # sample rows 
print('Number of NAs Per Column:\n' ,game_data.isna().sum()) # number of NAs per column 
```

```python

game_data.info()
game_data.describe()
game_data['home_won'].value_counts() 
game_data.head(5) #sample rows 
game_data.select_dtypes(['bool','str'])# there are two columns that are categorical, away_won and home_won within the dataset (bool values)
```

Important categorical columns in this dataset include the home_team, away_team, home_won, and away_won columns. The home and away team columns are important as they allow summary statistics to be calculated by franchise.
```python
# Histograms of columns 

cleaned_gamedata= game_data.drop(columns=['game_id']) # dropping the game ID column

axes = cleaned_gamedata.hist(figsize=(20, 20), bins=20)

for axe in axes.flatten(): 
    axe.set_ylabel('Frequency')
    axe.set_xlabel('Values')
plt.suptitle('Histograms of Game Data Columns')
plt.show()



# Plotting Average Goals in Home Vs Away Games 

avg_goals = {'Home': game_data['home_goals'].mean(),
             'Away': game_data['away_goals'].mean()}

plt.figure(figsize=(16, 8))
plt.bar(avg_goals.keys(), avg_goals.values())
plt.title('Average Goals in Home Vs Away Games')
plt.ylim(0,4)
plt.ylabel('Average Goals')
plt.xlabel('Game Type')


# Home vs Away Goals BoxPlot

home_goals = (game_data['home_goals']).dropna()
away_goals = (game_data['away_goals']).dropna()
plt.figure(figsize=(16, 8))
plt.boxplot([home_goals, away_goals])
plt.xticks([1, 2], ['Home Goals', 'Away Goals'])
plt.title('Distribution of Home vs Away Goals')
plt.ylabel('Goals')
plt.show()


# Home vs Away Goal Difference BoxPlot

game_data['goal_diff'] = game_data['home_goals'] - game_data['away_goals']

game_data.boxplot(column='goal_diff', figsize=(16, 8))
plt.title('Distribution of Goal Difference (Home Goals - Away Goals)')
plt.ylabel('Goal Difference')
plt.show()
```

The histograms of numeric variables in the dataset show that the game statistics are concentrated within common ranges with some outliers of high values. For example, most games had between 1 and 4 goals while a small number of games had unusually high scoring. This indicates that extreme scoring may have happened in the timeframe, but most were in this typical scoring range. 


These charts show that home teams perform slightly better than away teams across the dataset. The bar chart shows that home teams scored an average of approximately 2.9 goals per game while away teams scored 2.62. This difference suggests a moderate at home advantage for hockey games. 


The box plot supports this pattern as well. The median number of home goals is 3 compared to the median of 2 away goals. This shows that the typical home team scored one more goal than the away team. The overall distribution is very similar, however home goals are slightly more spread overall. While the median goal difference is 0, the box plot shows that many games were competitive, and the average goal difference is approximately 0.28 in favor of the home team in the dataset. 

# 6.) Categorical Data Analysis and Charts 

```python
game_data_categorical = game_data.select_dtypes(['bool', 'str'])

home_games_by_team = game_data_categorical['home_team'].value_counts()
away_games_by_team = game_data_categorical['away_team'].value_counts()
count_of_quarter_home_games_settled_in = game_data_categorical['home_settled_in'].value_counts()
count_of_quarter_away_games_settled_in = game_data_categorical['away_settled_in'].value_counts()

games_won = game_data_categorical[(game_data_categorical['away_won'] == True) | 
                                  (game_data_categorical['home_won']==True)] # this is the total games won in the season


games_lost = game_data_categorical[(game_data_categorical['away_won'] == False) | 
                                  (game_data_categorical['home_won']==False)]

tied_games = game_data_categorical[
    (~game_data_categorical['home_won']) &
    (~game_data_categorical['away_won'])
]

home_ties = tied_games.groupby('home_team').size()
away_ties = tied_games.groupby('away_team').size()

ties_per_team = home_ties.add(away_ties, fill_value=0).astype(int)


# Stats Grouped By Team With New Columns Added 

games_won_home = games_won[games_won['home_won'] == True].groupby('home_team').size()
games_won_away = games_won[games_won['away_won']==True].groupby( 'away_team').size() # the count of home won games 
games_lost_home = games_won[games_won['home_won']==False].groupby ('home_team').size()
games_lost_away= games_won[games_won['away_won']==False].groupby( 'away_team').size()




total_games_played_per_team = (
    games_won_home
    .add(games_won_away, fill_value=0)
    .add(games_lost_home, fill_value=0)
    .add(games_lost_away, fill_value=0)
    .add(ties_per_team,fill_value=0)
)

# New Column Creations 
total_games_won_per_team = games_won_home + games_won_away
total_games_lost_per_team = games_lost_home + games_lost_away


average_home_goals_per_team = game_data.groupby('home_team')['home_goals'].mean()
average_away_goals_per_team = game_data.groupby('away_team')['away_goals'].mean()


# Adding New Columns to one Dataframe  
games_status_data_frame = pd.DataFrame({'Home Game Wins Count': games_won_home,
              'Away Games Win Count':games_won_away,
              'Games Lost Home':games_lost_home,
              'Games Lost Away':games_lost_away,
              'Total Games Won Per Team':total_games_won_per_team,
              'Total Games Lost Per Team':total_games_lost_per_team,
              'Total Games Tied Per Team': ties_per_team, 
              'Total Games Played Per Team':total_games_played_per_team,
              'Home Win Percentage':(games_won_home/(games_won_home + games_lost_home)*100).round(2),
              'Home Loss Percentage':(games_lost_home/(games_won_home + games_lost_home)*100).round(2),
              'Away Win Percentage': (games_won_away/(games_won_away + games_lost_away)*100).round(2),
              'Away Loss Percentage': (games_lost_away/(games_won_away + games_lost_away)*100).round(2),
              'Average Home Goals Per Team':average_home_goals_per_team,
              'Average Away Goals Per Team':average_away_goals_per_team, 
              'Average Goals Per Team': (
        average_home_goals_per_team.add(average_away_goals_per_team, fill_value=0) / 2
    ).round(2)
})

games_status_data_frame = games_status_data_frame.reset_index()
games_status_data_frame = games_status_data_frame.rename(columns={'index': 'Team Name'})

outcome_counts = pd.Series({
    'Home Wins': game_data['home_won'].sum(),
    'Away Wins': game_data['away_won'].sum(),
    'Ties': ((~game_data['home_won']) & (~game_data['away_won'])).sum()
})

outcome_counts.plot(kind='bar', figsize=(8,6), color=['green', 'red', 'gray'])
plt.title('Game Outcomes')
plt.xlabel('Outcome')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()


settled_counts = game_data['home_settled_in'].value_counts()

plt.figure(figsize=(8,6))
plt.bar(settled_counts.index, settled_counts.values, color=['steelblue', 'orange', 'green'])
plt.title('How Home Team Wins Were Settled')
plt.xlabel('Game Settlement Type')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

print(games_status_data_frame.head(10),games_status_data_frame.describe())

```

In the above categorical analysis of wins at home, away and ties, we can see that the number of home wins is greater than the number of away wins with ties representing the minority of outcomes. This suggests that teams may have an at home advantage in hockey when compared to away games. This could be due to a number of factors such as crowd involvement, travel demands, and playing environment. This also indicates that location may have a significant impact on the performance of a team in a given season. 


The game settlement chart shows how home games were settled across the dataset. Most games ended in regualtion while fewer finished in OT or a shootout. This suggests that although many games were competitive, most were still settled in regulation.

```python
# Feature Engineeting - Conversion By Team (Goals / Shots)


home_team_goals = game_data.groupby('home_team')['home_goals'].sum()
home_team_shots = game_data.groupby('home_team')['home_shots'].sum()
away_team_goals = game_data.groupby('away_team')['away_goals'].sum()
away_team_shots = game_data.groupby('away_team')['away_shots'].sum()

home_conversion = home_team_goals / home_team_shots
away_conversion = away_team_goals / away_team_shots

conversion_df = pd.DataFrame({
    'Home Conversion %': home_conversion,
    'Away Conversion %': away_conversion
}).round(4)



# Plots of Grouped Comparisons


plot_df = games_status_data_frame.sort_values('Total Games Played Per Team', ascending=False)

plot_df.plot(
    x='Team Name',
    y=['Total Games Won Per Team', 'Total Games Lost Per Team'],
    kind='barh',
    figsize=(20,8),
    width = 0.8,
    color =['green','red'])

plt.title('Wins vs Losses by Team')
plt.xlabel('Count')
plt.ylabel('Team')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


plot_df = games_status_data_frame.sort_values('Home Win Percentage', ascending=False)

plot_df.plot(
    x='Team Name',
    y=['Home Win Percentage', 'Away Win Percentage'],
    kind='bar',
    figsize=(16,8),
    width = 0.8,
    color =['green','red'])
plt.title('Home Win % vs Away Win % by Team')
plt.xlabel('Team')
plt.ylabel('Win Percentage')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()


plot_df = games_status_data_frame.sort_values('Total Games Played Per Team', ascending=False)
plot_df.plot(
    x='Team Name',
    y=['Total Games Won Per Team', 'Total Games Lost Per Team'],
    kind='bar',
    stacked=True,
    figsize=(16,8),
    width = 0.8,
    color = ['green','red'])
plt.title('Total Proportion of Games Won and Lost by Team')
plt.xlabel('Team')
plt.ylabel('Games')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()


top10 = games_status_data_frame.sort_values('Total Games Won Per Team', ascending=False).head(10)

plt.figure(figsize=(16,8))
plt.barh(top10['Team Name'], top10['Total Games Won Per Team'])
plt.xticks(rotation=45)
plt.title('Top 10 Teams by Total Wins')
plt.xlabel('Wins')
plt.ylabel('Team')
plt.tight_layout()
plt.show()


average_goals_per_team = games_status_data_frame.sort_values('Average Goals Per Team', ascending = False)

plt.figure(figsize=(16,8))
plt.bar(average_goals_per_team['Team Name'], average_goals_per_team['Average Goals Per Team'])
plt.xticks(rotation=90)
plt.title('Goal Average By Team')
plt.xlabel('Team')
plt.ylabel('Goal Average')
plt.tight_layout()
plt.show()


conversion_df = conversion_df.sort_values('Home Conversion %', ascending=False)

conversion_df.plot(
    kind='barh',
    figsize=(16,10),
    width=0.8,
    color=['green', 'orange']
)

plt.title('Home vs Away Shot Conversion by Team (Goals / Shots)')
plt.xlabel('Conversion %')
plt.ylabel('Team')
plt.xlim(0, 0.15)
plt.tight_layout()
plt.show()
```

In the above code, feature engineering was applied to group the data together and calculate statistics by team aggregation. For example, conversion rate was calculated by grouping by team and then calculating the goals / shots for home and away games in the dataset. This is useful in identifying the proportion of shots that were successful in making a goal for a team and directly ties into wins due to scoring.

# 9.) Trends Over Time 

```python
# Time Series Data Analysis

game_data['date_time_GMT'] =pd.to_datetime(game_data['date_time_GMT']).dt.strftime('%Y-%m-%d') # converting to just a date for the datetime
game_data['Game Year'] =pd.to_datetime(game_data['date_time_GMT']).dt.strftime('%Y') # Extracting the year from the date for analysis


game_data['Game Year'].min()


home_team_goals_by_year = game_data.groupby(['Game Year'])[['home_goals']].sum()
away_team_goals_by_year = game_data.groupby(['Game Year'])[['away_goals']].sum()
away_team_shots_by_year = game_data.groupby(['Game Year'])[['away_shots']].sum()
home_team_shots_by_year = game_data.groupby(['Game Year'])[['home_shots']].sum()

home_shots_scored = home_team_goals_by_year['home_goals'] / home_team_shots_by_year['home_shots']
away_shots_scored = away_team_goals_by_year['away_goals'] / away_team_shots_by_year['away_shots']




fig, axes = plt.subplots(1, 2, figsize=(29, 10))

home_team_goals_by_year.plot(kind='line', ax=axes[0], label='Home Goals', color = 'green')
away_team_goals_by_year.plot(kind='line', ax=axes[0], label='Away Goals',color = 'red')
axes[0].set_title('Home Vs Away Goals Per Year')
axes[0].set_xlabel('Game Year')
axes[0].set_ylabel('Goals')
axes[0].tick_params(axis='x', labelrotation = 45)
axes[0].set_yticks(range(0, 5000,200))
axes[0].set_ylim(0, 5000)
axes[0].legend()


home_team_shots_by_year.plot(kind='line', ax=axes[1], label='Home Shots', color = 'green')
away_team_shots_by_year.plot(kind='line', ax=axes[1], label='Away Shots',color = 'red')
axes[1].set_title('Home Vs Away Shots Per Year')
axes[1].set_xlabel('Game Year')
axes[1].tick_params(axis = 'x', labelrotation = 45)
axes[1].set_ylabel('Shots')
axes[1].set_yticks(range(0, 50000,2000))
axes[1].set_ylim(0, 50000)
axes[1].legend()


conversion_df = pd.DataFrame({
    'Home Conversion %': home_shots_scored,
    'Away Conversion %': away_shots_scored
}).sort_index()

conversion_df.plot(kind='line',
    figsize=(16,8),
    color=['green', 'orange'],
    marker='o'
)

plt.title('Home vs Away Shot Conversion by Year 2000-2020')
plt.xlabel('Game Year')
plt.ylabel('Conversion Rate')
plt.xticks(rotation=45)
plt.ylim(0, 0.12)

plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()



# Chicago Blackhawks and Boston Bruins Total Wins By Year

selected_games = game_data[
    game_data['home_team'].isin(['Chicago Blackhawks', 'Boston Bruins','San Jose Sharks','Detroit Red Wings']) |
    game_data['away_team'].isin(['Chicago Blackhawks', 'Boston Bruins','San Jose Sharks','Detroit Red Wings'])]

home_wins = selected_games.loc[
    selected_games['home_won'],
    ['home_team', 'Game Year']].rename(columns={'home_team': 'Team'}) # renaming to join the DF together and count a win as a win regardless of home or away

away_wins = selected_games.loc[
    selected_games['away_won'],
    ['away_team', 'Game Year']].rename(columns={'away_team': 'Team'}) # renaming to join the DF together and count a win as a win regardless of home or away

all_wins = pd.concat([home_wins, away_wins])

all_wins = all_wins[
    all_wins['Team'].isin(['Chicago Blackhawks', 'Boston Bruins','San Jose Sharks', 'Detroit Red Wings'])]

wins_by_year = all_wins.groupby(['Game Year', 'Team']).size().unstack(fill_value=0)

wins_by_year.plot(kind='line', marker='o', figsize= (16,8), color = ['gold','red','hotpink','teal'])
plt.title('Wins by Year: Chicago Blackhawks, Boston Bruins, and San Jose Sharks')
plt.xlabel('Game Year')
plt.ylabel('Number of Wins')
plt.xticks(rotation=45)
plt.legend(title='Team')
plt.tight_layout()
plt.show()


```

In the above plots, time series trends were calculated for various scoring metrics. Total home and away goals, home and away shots, conversion rates by year, and team performance for a subset of teams were also plotted over time. 

The new feature conversion rate was utilized (calculated by dividing the number of goals by the number of attempts), to identify scoring patterns for each year. This feature is useful as it shows scoring efficiency rather than volume, and gives insight to the trends related to home vs away performance of teams. 

Together, these time-based plots provide a broader view of performance trends by year in the dataset. The goal and shot count charts display changes in offensive activity, while the conversion shows efficiency of scoring by year.

# LINEAR MODEL

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
import matplotlib.pyplot as plt


# Comparing the average home goals per team and the average away goals per team

model_df = games_status_data_frame[
    ['Average Home Goals Per Team', 'Average Away Goals Per Team']
].dropna()

X = model_df[['Average Home Goals Per Team']]
y = model_df['Average Away Goals Per Team']

model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)


# scatter plot of actual vs predicted values 
plt.scatter(y, y_pred, color = 'red')
plt.title('Actual vs Predicted Values of Average Goals')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')

print('Prediciton of Away Goal Average Based on Home Goals Average:')
print("Intercept:", round(model.intercept_, 4))
print("Coefficient:", round(model.coef_[0], 4))
print("R^2:", round(r2_score(y, y_pred), 4))
print("MSE:", round(mean_squared_error(y, y_pred), 4))

plt.figure(figsize=(8,6))
plt.scatter(X, y, alpha=0.7)
plt.plot(X, y_pred, color='red', linewidth=2)
plt.xlabel("Average Home Goals Per Team")
plt.ylabel("Average Away Goals Per Team")
plt.title("Linear Regression: Average Home Goals vs Average Away Goals")
plt.tight_layout()
plt.show()


# comparing the average home win percentage to the average away percentage. 

model_df = games_status_data_frame[
    ['Home Win Percentage', 'Away Win Percentage']
].dropna()

X = model_df[['Home Win Percentage']]
y = model_df['Away Win Percentage']

model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

print('Prediciton of Away Win % Based on Home Win %:')
print("Intercept:", round(model.intercept_, 4))
print("Coefficient:", round(model.coef_[0], 4))
print("R^2:", round(r2_score(y, y_pred), 4))
print("MSE:", round(mean_squared_error(y, y_pred), 4))


plt.figure(figsize=(8,6))
plt.scatter(y, y_pred, color='purple', alpha=0.7)
plt.xlabel('Actual Away Win Percentage')
plt.ylabel('Predicted Away Win Percentage')
plt.title('Actual vs Predicted: Away Win Percentage')

plt.tight_layout()
plt.show()


plt.figure(figsize=(8,6))
plt.scatter(X, y, alpha=0.7, color='purple')
plt.plot(X, y_pred, color='red', linewidth=2)
plt.xlabel("Home Win Percentage")
plt.ylabel("Away Win Percentage")
plt.title("Linear Regression: Home Win Percentage vs Away Win Percentage")
plt.tight_layout()
plt.show()
```
