Some other methods: flatten and ravel 

```python
x = np.arange(24).reshape(4,6)

x [0,0] = 99


y = np.ravel(x)

y[0]= 100 

print(f'y = \n {y}')
print(f'x= \n {x}')


z = (x).flatten() # flatten is creating a whole new array n one dimension 
z[-1] = 42

print(f'y = \n {y}')
print(f'x= \n {x}')
print(z)
```
