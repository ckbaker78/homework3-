Some other methods: flatten and ravel 

```python

import numpy as np 

x = np.arange(24).reshape(4,6) # This aranges a data frame from 0 to 23 

x [0,0] = 99


y = np.ravel(x)

y[0]= 100 

print(f'y = \n {y}')
print(f'x= \n {x}')


z = (x).flatten() # flatten is creating a whole new array n one dimension 
z[-1] = 42

print(f'y = \n {y}')
print(f'x= \n {x}')
print(f'z= \n {z}')
```

```python
# plotting with pandas 



import pandas as pd 

def palindrome_check(word): 

    if word[::1] == word[::-1]: 

        print('This is a palidrome!')

    else:
        print('This is not a palidrome!')




```

```python
a = [ 3, 4, 6, 9, 10, 12, 14, 15, 17, 19, 21 ]



def f(a,x): 

    valid = [i for i in a if i <= x or i == x]

    invalid = [i for i in a if x < i]

    for i in valid: 

        if i > x or i < x:
            return np.max(valid)
        
    for i in invalid: 

        return -1 

result = f(a,1)

print(result)

```

```python
def fibonacci(n):
    result = [0,1]


    for i in range(n-2):
        result.append(result[-1] + result[-2])

    return result

print(fibonacci(10))
```
