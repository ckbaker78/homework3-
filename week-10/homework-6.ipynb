**Task:**

Write a function `read_and_plot_column(filename, column_name)` that:
* Opens a CSV file (do **not** use pandas)
* Extracts the values from the specified column
* Casts all values to `float`
* Plots the values using `plt.plot()`
* Adds a title
* Displays the plot

````python
# Your Solution

import csv


def read_and_plot_column(filename,column_name):
    '''Reding a CSV and plotting column name ''' 

    values = []


    with open(filename, 'r') as f: 

        reader = csv.DictReader(f)

        for row in reader: 

            values.append(float(row[column_name]))

    plt.plot(values)
    plt.title(f"{column_name} from {filename}")
    plt.ylabel('Temp')
    plt.show()
````

Using `df_weather`:
* Write a helper function to extract time (in hours) from a datetime string and convert it to decimal hours (float)
* Convert time using:

    `hours + (minutes / 60) + (seconds / 3600)`
* Create a new DataFrame with the same index as `df_weather`
* Add the following columns:
  * `sunrise` (converted to float)
  * `sunset` (converted to float)
  * `tempmin` (copied directly from `df_weather`)
* Plot all columns together using `.plot()`


**Sample Output:**

This is similar to how your new dataframe should look before plotting:

```
   start_time  end_time  price
P1       8.25     17.50     20
P2       8.50     17.75     35
P3       9.00     18.00     15
P4       9.25     18.25     60
P5       9.50     18.50     50
```

````python
# Your Solution

def time_helper(datecolumn): 
    '''This will take the datecolumn and split on T
    Then it will take the 2nd half of the split(after T)
    it will then split on a colon to separate the other parts of time'''

    x = datecolumn.split('T')[1] # taking the second half of the split on T values 

    h,m,s = x.split(':')

    return float(h) + float(m)/60 + float(s)/3600


df_new = pd.DataFrame()

# add columns
df_new['sunrise'] = df_weather['sunrise'].apply(time_helper)
df_new['sunset'] = df_weather['sunset'].apply(time_helper)
df_new['tempmin'] = df_weather['tempmin']


print(df_new)
````

* Returns the updated DataFrame

**Sample Output:**

```

P1     20         2     18
P2     35         5     30
P3     15         1     14
P4     60        10     50
P5     50         8     42
```

````python
# Your Solution


def add_range_column(df,col_a,col_b): 

    '''This converts the desired columns into a float via the .astype() method and then calculates the range. 
    It then inserts the column Range into the dataframe as a new column'''

    col_a = df[col_a].astype(float)
    col_b = df[col_b].astype(float)

    df['range'] = col_a - col_b

    return df
````

**Problem 17**

**Concept:** *Filtering records by category.*


**Task:**

Write a function `filter_weather(df, conditions, temp_min, temp_max)` that:
* Filters rows where:
  * conditions contain the given string
  * `tempmax` ≤ `temp_max`
  * `tempmin` ≥ `temp_min`
* Prints intermediate counts
* Returns the filtered DataFrame

**Sample Output:**

If your data matched the filtering conditions step by step, the following could be your output:
```
4 records with conditions Blue
3 records with conditions Blue and tempmax <= 50
2 records with conditions Blue and 20 <= temp <= 50
Found 2 records with conditions of Blue and temp between 20 and 50.
```

````python
# Your Solution


def filter_weather(df, conditions, temp_min, temp_max):
    case1 = df[df['conditions'].str.contains(conditions, na=False)]
    print(f"{len(case1)} records with conditions {conditions}")

    case2 = case1[case1['tempmax'] <= temp_max]
    print(f"{len(case2)} records with conditions {conditions} and tempmax <= {temp_max}")

    case3 = case2[case2['tempmin'] >= temp_min]
    print(f"{len(case3)} records with conditions {conditions} and {temp_min} <= temp <= {temp_max}")

    print(f"Found {len(case3)} records with conditions of {conditions} and temp between {temp_min} and {temp_max}.")
    return case3

````

**Concept:** *Creating DataFrame, standardizing values.*


**Task:**

Write a function `create_standardized_values(df_source, columns, row_index=None)` that:
* Creates a new DataFrame
* Standardizes values relative to a given row
* Uses the first row if none is provided
* Returns the new DataFrame

**Sample Output:**

If the following was your dataframe:
```
      high_value  low_value
2020        100        50
2021        120        55
2022        140        60
2023        160        70
2024        180        75
```
If you call the function
`create_standardized_values(df_example, ["high_value", "low_value"], 2020)`,
you should get the following as the output:
```
      high_value  low_value
2020         1.0       1.00
2021         1.2       1.10
2022         1.4       1.20
2023         1.6       1.40
2024         1.8       1.50

```

````python
# Your Solution

def create_standardized_values(df_source, columns, row_index=None):

    if row_index is None:
        row_index = df_source.index[0]
    
    df_new = pd.DataFrame(index=df_source.index)
    
    for col in columns:
        baseline = df_source.loc[row_index, col] # loc pulls the rows based on an index or column label. 
        df_new[col] = df_source[col] / baseline
    
    return df_new

````
