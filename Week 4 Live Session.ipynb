```python
# Accumulator functions 

def add_up(numbers): 
     
     # declare an accumulator
     # loop over sequence 
     # update accumulator inside loop 
    total = 0 

    for n in numbers: 
        total += n # total + n is also equal to this 
    return total 


add_up(numbers)
print('result =', add_up(numbers))
```

```python
def factorial(n):
    '''return the factorial of n
fact(n) n* (n-1) * (n-2 )...'''

    fact = 1

    for i in range(n, 0, -1): # this reads as starting at n, stopping before 0, and counting down (-1)

       # print(f'i = {i}') # this tests the statement 
        fact = fact * i 
    
    return fact


print(factorial(4))


```

```python
# The same as above, but now producing a list

def reciprocal_numbers(n): 
    '''return a list of numbers [1, 1/1+1, 1/1+2...]''' 

    result = []

    for i in range(n): # counting the integer numbers in up to N values 

        r = 1 / (i+1)

        print(f'i = {i}, r = {r}')

        result.append(r)

        #result = result + [r] # if you wrap r in brackets, this is equivalent to append. 


    return result # this returns the list of numbers that we appended into the list 


#### We use append when wanting to return a list from the function #### 


reciprocal_numbers(5)

print('reciprocal numbers(5) = ', reciprocal_numbers(5)) 
```

```python
def fibonacci(n): 

    '''return n fibonacci numbers
    [0,1,1,2,3,5,8,13]'''

    total = [0,1] # starting at 0 and one total[0] = 0 total[1] = 1 

    for i in range(2,n): # starting at sequence to to n 

        fib = total[-1] + total[-2] # this is taking the right most and 2nd right most postions 

        total.append(fib)

    return total 


fibonacci(10)
```

```python
# ASCII Example 

# ord() is used to convert strings to ascii 

# chr() is used to convert code to strings 


def ascii_conversion(n): 

    total = []

    for i in n: 
        ascii = ord(i)

        total.append(ascii)

    return total


def ascii_to_string(n):

    result = ''

    for i in n:
        string = chr(i)
        result += string # creating a whole new sting through concatenation 


    return result

ascii_to_string(ascii_conversion('hello'))


```

```python
# The Cesar Cypher Encyrption # 

def cipher(s,n): 

    string = ''


    for i in s: 
        code = ord(i)

        shift = code + n

        string = string + chr(shift)
    return string 


cipher('hello', 1)
```
